```python
day_by = df123.groupby('시간대').sum()['통화건수'].reset_index(name="통화건수 합")
day_by
```

```python
import datetime
import random
import matplotlib.pyplot as plt

# make up some data
x = list(day_by['시간대'])
y = list(day_by['통화건수 합'])

# plot
plt.plot(x,y)
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```
